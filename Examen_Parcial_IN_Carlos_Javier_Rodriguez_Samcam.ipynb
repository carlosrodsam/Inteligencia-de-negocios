```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/carlosrodsam/spotify_dataset/main/cfb20.csv'
df=pd.read_csv(url)
df
```

```python
df.set_index('Def.Rank')
fig = plt.figure()
x1 = fig.add_subplot(111)
df['Off.Yards.Play'].plot(ax=x1,label='Off.Yards.Play')
plt.xlabel('Def.Rank')
x1.set_ylabel('Off.Yards.Play')
plt.legend(loc=1)

x2 = x1.twinx()

df['Off.Yards.per.Game'].plot(ax=x2, label = 'Off.Yards.per.Game', color='green')
x2.set_ylabel('Off.Yards.per.Game')
plt.legend(loc=1, bbox_to_anchor=(1,0.9))


plt.title('Def.Rank based on off.yards play')

plt.show()
```

JustificaciÃ³n de selecciÃ³n: Lo use para ver similitudes entre "Off Yard Play" y "Off Yards Per Game" y que tanto variaban con forme a Def. Rank

ConclusiÃ³n: No existe una variacion marcada de los datos mostrados en relacion a Def.Rank y no afecta mucho en cuanto aumenta, lo que si se puede observar es que entre la mediana de los Def Rank los Off Yard Play y OFf Yards per Game tienen un aumento significante

```python
url2 = 'https://raw.githubusercontent.com/carlosrodsam/spotify_dataset/main/cbb21.csv'
df2=pd.read_csv(url2, index_col=0)
df2
```

```python
df2 = df2.drop(df2.columns[[0,1]], axis=1)
df2.index.TEAM = ['TEAM']

fig, ax = plt.subplots(figsize=(11, 5))
hvst = sns.heatmap(df2, cmap="Blues")
hvst.set_yticklabels(hvst.get_yticklabels(), rotation=0)

plt.tight_layout()
plt.show()
```
